# BLS Data Download — bulk flat files

```python
import logging
import sys
from datetime import date

logging.basicConfig(level=logging.INFO, stream=sys.stderr)

start_date = date(2024, 4, 1)
end_date = date(2024, 6, 1)
```

## QCEW (no API key required)

```python
from bls_stats.download.qcew import download_qcew

qcew_df = download_qcew(start_date, end_date)
print(f"QCEW: {len(qcew_df)} rows")
qcew_df.head()
```

### BED

```python
from bls_stats.download.bed import download_bed

bed_df = download_bed(start_date, end_date)
print(f"BED: {len(bed_df)} rows")
bed_df.head()
```

### CES

```python
from bls_stats.download.ces import download_ces

ces_df = download_ces(start_date, end_date)
print(f"CES: {len(ces_df)} rows")
ces_df.head()
```

### SAE

```python
from bls_stats.download.sae import download_sae

sae_df = download_sae(start_date, end_date)
print(f"SAE: {len(sae_df)} rows")
sae_df.head()
```

### JOLTS

```python
from bls_stats.download.jolts import download_jolts

jolts_df = download_jolts(start_date, end_date)
print(f"JOLTS: {len(jolts_df)} rows")
jolts_df.head()
```

## Summary

```python
import polars as pl

summary = pl.DataFrame({
    "program": ["QCEW", "BED", "CES", "SAE", "JOLTS"],
    "rows": [len(qcew_df), len(bed_df), len(ces_df), len(sae_df), len(jolts_df)],
})
summary
```
