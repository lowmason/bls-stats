```python
import logging
import sys

logging.basicConfig(level=logging.INFO, stream=sys.stderr)

monthly_periods = [(2024, 4), (2024, 5), (2024, 6)]
quarterly_periods = [(2024, 2)]
```

```python
from bls_stats.download.qcew import download_qcew

qcew_df = download_qcew(quarterly_periods)
print(f"QCEW: {len(qcew_df)} rows")
qcew_df.head()
```

```python
from bls_stats.download.bed import download_bed

bed_df = download_bed(quarterly_periods)
print(f"BED: {len(bed_df)} rows")
bed_df.head()
```

```python
from bls_stats.download.ces import download_ces

ces_df = download_ces(monthly_periods)
print(f"CES: {len(ces_df)} rows")
ces_df.head()
```

```python
from bls_stats.download.sae import download_sae

sae_df = download_sae(monthly_periods)
print(f"SAE: {len(sae_df)} rows")
sae_df.head()
```

```python
from bls_stats.download.jolts import download_jolts

jolts_df = download_jolts(monthly_periods)
print(f"JOLTS: {len(jolts_df)} rows")
jolts_df.head()
```

```python
import polars as pl

summary = pl.DataFrame({
    "program": ["QCEW", "BED", "CES", "SAE", "JOLTS"],
    "rows": [len(qcew_df), len(bed_df), len(ces_df), len(sae_df), len(jolts_df)],
})
summary
```

```python
for name, df in [
    ('qcew', qcew_df),
    ('bed', bed_df),
    ('ces', ces_df),
    ('sae', sae_df),
    ('jolts', jolts_df)
]:
    print(f'bls_stats_{name}')
    for col, dtype in df[1].schema.items():
        print(f"{col} - {dtype}")
    print()
```

```python
from bls_stats.bls.periods import reference_periods

print("=== Monthly programs (year, month) ===\n")

for prog in ["ces", "sae", "jolts"]:
    periods = reference_periods(prog, "2024/01", "2024/06")
    print(f"{prog.upper():>5s}: {periods}")

print("\n=== Quarterly programs (year, quarter) ===\n")

for prog in ["qcew", "bed"]:
    periods = reference_periods(prog, "2024/1", "2024/4")
    print(f"{prog.upper():>5s}: {periods}")

print("\n=== Cross-year boundary (JOLTS 2023/10 → 2024/03) ===\n")
print(reference_periods("jolts", "2023/10", "2024/03"))
```
